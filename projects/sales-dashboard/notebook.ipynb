```python
import pandas as pd
import matplotlib.pyplot as plt

# simulated data
df = pd.DataFrame({
    'month': ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'],
    'sales': [120,180,150,220,260,300,280,310,340,320,360,400]
})

print(df.head())
plt.figure(figsize=(8,3))
plt.bar(df['month'], df['sales'], color='#0ea5a6')
plt.title('Monthly Sales (simulated)')
plt.tight_layout()
plt.show()
```
